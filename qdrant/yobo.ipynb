***Base-64 image conversion***

```python
from PIL import Image
from io import BytesIO
import base64

def pil_image_to_base64(pil_image):
    """
    Convert a PIL image to a base64-encoded string.

    Args:
        pil_image (PIL.JpegImagePlugin.JpegImageFile): A PIL image object.

    Returns:
        str: The base64-encoded image data.
    """
    byte_io = BytesIO()
    pil_image.save(byte_io, format='JPEG')
    byte_image = byte_io.getvalue()
    base64_image = base64.b64encode(byte_image).decode('utf-8')
    return base64_image
```

***Embedding***

```python
from sentence_transformers import SentenceTransformer
image_model = SentenceTransformer('clip-ViT-B-32')
```

```python
import requests
from pathlib import Path

output_directory = Path("./imagesall")
output_directory.mkdir(exist_ok=True)

iteration_cnt = 0
for index, row in product_df.iterrows():
    iteration_cnt = iteration_cnt + 1
    if iteration_cnt < 13378:
      continue
    summary_str = "This is a " + row['Category'] + " from " + row['Company'] +". It is a " + row['Name'] + ". " + row['Description']
    name_embedding = openai_client.embeddings.create(input= row['Name'], model=embedding_model)
    description_embedding = openai_client.embeddings.create(input= row['Description'], model=embedding_model)
    specifications_embedding = openai_client.embeddings.create(input= row['Specifications'], model=embedding_model)
    summary_embedding = openai_client.embeddings.create(input= summary_str, model=embedding_model)

    image_url = row['Image_links'][0]
    image_filename = image_url.split('/')[-1].split('?')[0]
    image_path = output_directory / image_filename

    if not image_path.exists():
      response = requests.get(image_url)
      if response.status_code == 200:
        with open(image_path, 'wb') as f:
            f.write(response.content)
      else:
        print(f"Failed to download image from {image_url}")
    else:
      print(f"Image already exists at {image_path}")

    img = Image.open(image_path)
    query_embedding = image_model.encode(img)
    finl_img_embedding = query_embedding.tolist()
    # if len(finl_img_embedding) != 512:
    #   print("error", image_path)


    # print("ok")
    # print(len(name_embedding.data[0].embedding))

    client.upsert(
        collection_name="products",
        points=[
            {
                "id": index,
                "vector": {
                    "name": name_embedding.data[0].embedding,
                    "description": description_embedding.data[0].embedding,
                    "specifications": specifications_embedding.data[0].embedding,
                    "summary": summary_embedding.data[0].embedding,
                    "image": finl_img_embedding
                },
                "payload": {
                    "product_id": row["product_id"],
                    "Link": row["Link"],
                    "Name": row["Name"],
                    "Description": row["Description"],
                    "Price": row["Price"],
                    "Specifications": row["Specifications"],
                    "Image_links": row["Image_links"].tolist(),
                    "Gender": row["Gender"],
                    "Category": row["Category"],
                    "Company": row["Company"],
                }
            }
        ]
    )
    print(row["product_id"], " done")


```
